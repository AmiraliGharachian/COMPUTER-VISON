```python

import cv2
import matplotlib.pyplot as plt
import requests
import numpy as np

# تابع برای محاسبه میزان روشنایی تصویر
def calculate_brightness(image):
    # تبدیل تصویر به حالت سیاه و سفید
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    # محاسبه میانگین روشنایی پیکسل‌ها
    brightness = int(gray_image.mean())
    return brightness


# لینک تصویر را وارد کنید
image_url = "https://www.cdc.gov/importation/images/cat.jpg?_=18560"

# دریافت تصویر از لینک
response = requests.get(image_url, stream=True).raw
# تبدیل تصویر به آرایه نامپای
image = np.asarray(bytearray(response.read()), dtype="uint8")
# خواندن تصویر با استفاده از OpenCV
image = cv2.imdecode(image, cv2.IMREAD_COLOR)
# محاسبه میزان روشنایی تصویر
brightness = calculate_brightness(image)

# تبدیل تصویر به حالت RGB
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# پردازش تصویر و محاسبه هیستوگرام روشنایی
histogram = cv2.calcHist([image], [0], None, [256], [0, 256])

# تنظیمات نمودار
plt.style.use('ggplot')
fig, ax = plt.subplots(figsize=(8, 5))

# رسم نمودار هیستوگرام با تنظیمات مناسب
ax.plot(histogram, color='gray', alpha=0.7)
ax.set_title("Histogram of Brightness")
ax.set_xlabel("Brightness")
ax.set_ylabel("Frequency")

# تنظیمات محور افقی (X)
ax.set_xlim(0, 255)
ax.set_xticks(np.arange(0, 256, 32))
ax.set_xticklabels(np.arange(0, 256, 32))

# تنظیمات محور عمودی (Y)
ax.set_ylim(0, np.max(histogram) + 1000)
ax.set_yticks(np.arange(0, np.max(histogram) + 1000, 5000))

# نمایش شبکه‌های راهنمای نمودار
ax.grid(True, linestyle='--', alpha=0.5)

# نمایش تصویر و نمودار
fig.tight_layout()
plt.show()


```

```python
# نمایش نویز تصویر
import cv2
import numpy as np
from matplotlib import pyplot as plt

# تابع برای دریافت تصویر از لینک
def get_image_from_url(url):
    resp = urllib.request.urlopen(url)
    image = np.asarray(bytearray(resp.read()), dtype="uint8")
    image = cv2.imdecode(image, cv2.IMREAD_GRAYSCALE)
    return image

# لینک تصویر
image_url = "https://www.cdc.gov/importation/images/cat.jpg?_=18560"

# دریافت تصویر از لینک
image = get_image_from_url(image_url)

# اعمال تبدیل فوریه
fft = np.fft.fft2(image)
fft_shifted = np.fft.fftshift(fft)
magnitude_spectrum = 20 * np.log(np.abs(fft_shifted))

# نمایش تصویر فرکانس
plt.imshow(magnitude_spectrum, cmap='gray')
plt.title('Magnitude Spectrum')
plt.show()

```

```python
# الگوریتم حذف نویز
import cv2
import numpy as np
from matplotlib import pyplot as plt
import urllib.request

# تابع برای دریافت تصویر از لینک
def get_image_from_url(url):
    resp = urllib.request.urlopen(url)
    image = np.asarray(bytearray(resp.read()), dtype="uint8")
    image = cv2.imdecode(image, cv2.IMREAD_GRAYSCALE)
    return image

# تابع برای کاهش نویز تصویر با استفاده از تبدیل فوریه
def denoise_image(image):
    # اعمال تبدیل فوریه
    fft = np.fft.fft2(image)
    fft_shifted = np.fft.fftshift(fft)

    # حذف فرکانس‌های با انرژی کم (مثلاً 10 درصد بالا و 10 درصد چپ)
    rows, cols = image.shape
    center_row, center_col = int(rows / 2), int(cols / 2)
    percentage = 10  # درصد فرکانس‌های با انرژی کم
    fft_shifted[:center_row - int(center_row * percentage / 100), :] = 0
    fft_shifted[center_row + int(center_row * percentage / 100):, :] = 0
    fft_shifted[:, :center_col - int(center_col * percentage / 100)] = 0
    fft_shifted[:, center_col + int(center_col * percentage / 100):] = 0

    # اعمال تبدیل معکوس فوریه
    ifft_shifted = np.fft.ifftshift(fft_shifted)
    restored_image = np.abs(np.fft.ifft2(ifft_shifted))

    # محدود کردن مقدار پیکسل‌ها به محدوده صحیح (0 تا 255)
    restored_image = np.clip(restored_image, 0, 255)
    restored_image = np.uint8(restored_image)

    return restored_image

# لینک تصویر
image_url = "https://www.cdc.gov/importation/images/cat.jpg?_=18560"

# دریافت تصویر از لینک
image = get_image_from_url(image_url)

# کاهش نویز تصویر با استفاده از تبدیل فوریه
denoised_image = denoise_image(image)

# نمایش تصویر قبل و بعد از کاهش نویز
plt.subplot(1, 2, 1), plt.imshow(image, cmap='gray')
plt.title('Noisy Image'), plt.xticks([]), plt.yticks([])
plt.subplot(1, 2, 2), plt.imshow(denoised_image, cmap='gray')
plt.title('Denoised Image'), plt.xticks([]), plt.yticks([])
plt.show()

```
